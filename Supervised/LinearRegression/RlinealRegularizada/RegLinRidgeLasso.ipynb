```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = 'Consumo de Combustible CO2.xlsx'
df = pd.read_excel(data)
df
```

```python
df.info()   # No existen valores nulos en este dataframe
```

```python
# Remover caracteristicas categoricas
df.drop(['MODELYEAR', 'MAKE', 'MODEL', 'VEHICLECLASS', 'ENGINESIZE', 'CYLINDERS', 'TRANSMISSION', 'FUELTYPE'], axis = 1, inplace = True)
df
```

```python
x = df.drop('CO2EMISSIONS', axis = 1)
y = df['CO2EMISSIONS']
```

```python
x
```

```python
y
```

```python
# Division en los conjuntos de prueba y de entrenamiento
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.30, random_state = 1)
```

# Regression Lineal multiple

```python
# Construccion de un modelo de regresion lineal multiple
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
linreg.fit(x_train, y_train)
```

```python
# Impresion de los coeficientes de regresiÃ³n 
print('Intercepto:', linreg.intercept_)
print('Coeficientes:', linreg.coef_)
```

```python
# Predicciones
y_pred = linreg.predict(x_test)

# Calculo de indicadores de la bondad de ajuste
from sklearn.metrics import r2_score
from sklearn import metrics
print('Valor de R cuadrada:', r2_score(y_test, y_pred))
print('Error absoluto medio:', metrics.mean_absolute_error(y_test, y_pred))
print('Error cuadratico medio:', metrics.mean_squared_error(y_test, y_pred))
print('Raiz del error cuadratico medio:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

# Regression Ridge con 

```python
from sklearn.linear_model import RidgeCV
# Definir un rango de prueba para alpha
alpha_range = 10.**np.arange(-3,3)

# normalizacion de los datos por que el parametro normalize ya no esta disponible en esta version
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)

ridgeregcv = RidgeCV(alphas = alpha_range, scoring = 'neg_mean_squared_error')
ridgeregcv.fit(x_train_scaled, y_train)
ridgeregcv.alpha_
```

```python
# examinacion de los coeficientes de la regresion RidgeCV
print('Intercepto:', ridgeregcv.intercept_)
print('Coeficientes:', ridgeregcv.coef_)
```

```python
# Prediccion mediante el mejor valor de alpha
y_pred = ridgeregcv.predict(x_test_scaled)

# Impresion de indicadores de bondad de ajuste
print('Valor de R cuadrada:', r2_score(y_test, y_pred))
print('Error absoluto medio:', metrics.mean_absolute_error(y_test, y_pred))
print('Error cuadratico medio:', metrics.mean_squared_error(y_test, y_pred))
print('Raiz del error cuadratico medio:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

# Regression Lasso con Alpha optimo 

```python
# Seleccion del valor optimo de Alpha para la regression de lasso
from sklearn.linear_model import LassoCV
lassoregcv = LassoCV(n_alphas = 100, random_state = 1)
lassoregcv.fit(x_train_scaled, y_train)
print('Alpha optimo:', lassoregcv.alpha_)
```

```python
print('Intercepto:', lassoregcv.intercept_)
print('Coeficientes:', lassoregcv.coef_)
```

```python
# Prediccion mediante regresion de lasso con un Alpha optimo
y_pred = lassoregcv.predict(x_test_scaled)

# Impresion de indicadores de bondad de ajuste
print('Valor de R cuadrada:', r2_score(y_test, y_pred))
print('Error absoluto medio:', metrics.mean_absolute_error(y_test, y_pred))
print('Error cuadratico medio:', metrics.mean_squared_error(y_test, y_pred))
print('Raiz del error cuadratico medio:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
